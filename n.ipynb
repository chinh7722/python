```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import threading
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
import time
from selenium.webdriver.chrome.options import Options
import pandas as pd
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# Using Chrome browser
driver = webdriver.Chrome()
```

```python
url = 'https://books.toscrape.com/'
driver.get(url)  
```

```python
search = driver.find_element(By.XPATH, value='//*[@id="APjFqb"]')
print(search)
```

```python
search.send_keys("python" + Keys.ENTER)
```

```python
books=driver.find_elements(By.CLASS_NAME, value='col-xs-6')
print(len(books))
```

```python
data= {
    'số thứ tự':[],
    'tên sách':[]
}
i=1
for book in books:
    ten=book.find_element(By.XPATH, value='.//h3').text
    data['số thứ tự'].append(i)
    data['tên sách'].append(ten)
    i+=1
```

```python
links=[]
for book in books:
    link=book.find_element(By.XPATH, value='.//h3/a').get_attribute("href")
    links.append(link)
```

```python
data= {
    'số thứ tự':[],
    'tên sách':[]
}
i=1
for link in links:
    driver.get(link)
    #chờ web load
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.content")))
    ten=driver.find_element(By.XPATH, value='//*[@id="content_inner"]/article/div[1]/div[2]/h1').text
    data['tên sách'].append(ten)
    data['số thứ tự'].append(i)
    i+=1
```

```python
print(data)
```

```python
df = pd.DataFrame(data)

print("DataFrame đã tạo:") 
print(df)
```

```python
df.to_csv('ten_truyen.csv', index=False, encoding='utf-8')   
```
